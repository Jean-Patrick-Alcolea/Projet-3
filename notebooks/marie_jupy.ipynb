```python
import requests
from bs4 import BeautifulSoup
import re

url = 'https://france.comersis.com/departement.php'
html = requests.get(url)

soup = BeautifulSoup(html.text, 'html.parser')
list_a = soup.find_all('a')

liens = []
for li in list_a:
    if re.search(r'departement\.\w+\?\w+\=\w+', str(li)):
        liens.append(re.search(r'departement\.\w+\?\w+\=\w+', str(li)).group(0))
liens.append('departement.php?dpt=75')
```

```python
def list_all_months_recol(cells):
    # Dictionnaire pour convertir le nom du mois en sa position dans l'année
    month_order = {
        "janvier": 1,
        "février": 2,
        "mars": 3,
        "avril": 4,
        "mai": 5,
        "juin": 6,
        "juillet": 7,
        "août": 8,
        "septembre": 9,
        "octobre": 10,
        "novembre": 11,
        "décembre": 12
    }

    # Liste pour stocker tous les mois
    all_months = []

    if cells == "toute l'année":
        return list(month_order.keys())
    else:
        start_month, end_month = cells.split(" à ")
        start = month_order[start_month]
        end = month_order[end_month]

        # Ajouter tous les mois de la plage à la liste
        if start <= end:
            # Cas normal où la plage ne traverse pas la fin de l'année
            for month_num in range(start, end + 1):
                for month_name, num in month_order.items():
                    if num == month_num:
                        all_months.append(month_name)
        else:
            # Cas où la plage traverse la fin de l'année
            for month_num in range(start, 13):
                for month_name, num in month_order.items():
                    if num == month_num:
                        all_months.append(month_name)
            for month_num in range(1, end + 1):
                for month_name, num in month_order.items():
                    if num == month_num:
                        all_months.append(month_name)

    return all_months
```

```python
from sklearn.neighbors import NearestNeighbors
from sklearn.feature_extraction.text import TfidfVectorizer
import pandas as pd

# Vectorisation des données textuelles

# Combiner toutes les colonnes en une seule string par ligne
X_text = df[['Nom','Sol requis', 'Sol requis','Sol requis', 'Zone climatique idéale','Zone climatique idéale','Zone climatique idéale','Type','Type','Type','Besoin en eau', 'Besoin en eau', 'Besoin en eau',
             'Rusticité','Rusticité','Rusticité','Exposition','Exposition','Calendrier des semis']].apply(
    lambda row: ' '.join(row.astype(str)), axis=1
)

tfidf = TfidfVectorizer()
X_vec = tfidf.fit_transform(X_text)

# Créer et entraîner le modèle k-NN
model = NearestNeighbors(n_neighbors=6)
model.fit(X_vec)

# Recherche des voisins les plus proches pour chaque plante
distances, indices = model.kneighbors(X_vec)

# Création d'un dictionnaire pour stocker les résultats
results = {}

for i in range(len(indices)):
    plante_reference = df.iloc[i]['Nom']
    nearest_neighbors = df.iloc[indices[i][1:]]['Nom'].tolist()  # [1:] pour exclure la première (elle-même)
    #nearest_neighbors = df.iloc[indices[i]]['Nom'].tolist()
    results[plante_reference] = nearest_neighbors

# Affichage des résultats
for plante, neighbors in results.items():
    print(f"Les 5 plantes les plus proches de {plante} sont : {neighbors}")
```
