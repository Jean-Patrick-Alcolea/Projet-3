```python
import pandas as pd
```

```python
df_plantes_photos = pd.read_csv('../data/plantes_photos.csv')
```

```python
df_plantes_photos.sample(10)
```

```python
df_plantes_photos.info()
```

```python
#Reset index
#Enelver cm profondeur et espacement
#enlever " °C" à Temperature
#enlever jours à delai_récolte
# temps_levée, temperature et delai_recole en int
#profondeur, espacement en float
```

```python
df_plantes_photos['Profondeur'] = df_plantes_photos['Profondeur'].str.replace(' cm', '')
df_plantes_photos['Espacement'] = df_plantes_photos['Espacement'].str.replace(' cm', '')
df_plantes_photos['Temps_levée'] = df_plantes_photos['Temps_levée'].str.replace(' jours', '')
df_plantes_photos['delai_recolte'] = df_plantes_photos['delai_recolte'].str.replace(' jours', '') #aneth à 3 ans, garder ou la passer en none ? Convertir en jours ? 
df_plantes_photos['Temperature'] = df_plantes_photos['Temperature'].str.replace(' °C', '')
#df_plantes_photos.drop(columns=['Unnamed: 0.1', 'Unnamed: 0'], inplace=True)
df_plantes_photos.sort_values(['Type', 'Nom'], inplace=True)
df_plantes_photos.reset_index(drop=True, inplace=True)
```

```python
round(df_plantes_photos['Type'].value_counts(normalize=True) * 100,2)
```

```python
df_plantes_photos['Espacement'] = df_plantes_photos['Espacement'].astype(float)
df_plantes_photos['Profondeur'] = df_plantes_photos['Profondeur'].astype(float)
df_plantes_photos['Temps_levée'] = df_plantes_photos['Temps_levée'].astype('Int64')
```
